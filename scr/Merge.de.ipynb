```python
import pandas as pd
# Erneute Definition der Dateipfade
time_series_path = "../data/time_series_60min_singleindex.csv"
weather_data_path = "../data/weather_data.csv"

# Einlesen mit Spaltenfilter
time_series_de = pd.read_csv(time_series_path, usecols=[
    "utc_timestamp",
    "DE_load_actual_entsoe_transparency",
    "DE_solar_generation_actual",
    "DE_wind_onshore_generation_actual"
], parse_dates=["utc_timestamp"])

weather_de = pd.read_csv(weather_data_path, usecols=[
    "utc_timestamp",
    "DE_temperature",
    "DE_radiation_direct_horizontal",
    "DE_radiation_diffuse_horizontal"
], parse_dates=["utc_timestamp"])

# Filterzeitraum 2015â€“2020
start_date = "2015-01-01"
end_date = "2020-12-31"

time_series_de = time_series_de[(time_series_de["utc_timestamp"] >= start_date) & (time_series_de["utc_timestamp"] <= end_date)]
weather_de = weather_de[(weather_de["utc_timestamp"] >= start_date) & (weather_de["utc_timestamp"] <= end_date)]

# ZusammenfÃ¼hren auf Basis von Zeit
merged_df = pd.merge(time_series_de, weather_de, on="utc_timestamp", how="inner")

#speichern als eine CSV-Datei
output_path = "../data/merged_data_de.csv"
merged_df.to_csv(output_path, index=False)  

#zeigen die ersten 5 Zeilen des DataFrames
print(merged_df.head(20))
```
